```python
import numpy as np
import pandas as pd
import umap
import seaborn as sns
```

```python
dataframe = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/iris.data')
```

```python
X = dataframe.drop(columns='Name')
y = dataframe['Name']
```

```python
reducer = umap.UMAP()
```

```python
data_reduced = reducer.fit_transform(X=X)
```

```python
coluns_name  = {0:'X',1:'Y'}
data_reduced = pd.DataFrame(data_reduced)
data_reduced.rename(columns=coluns_name,inplace=True)
```

```python
data_reduced['target'] = y
```

```python
data_reduced
```

```python
sns.scatterplot(x='X', y='Y', hue='target', data=data_reduced, palette='viridis', s=70)
```
