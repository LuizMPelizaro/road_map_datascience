```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro
import scipy.stats as stats
import numpy as np
```

```python
dataframe = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/iris.data')
dataframe.drop(columns=['Name'],inplace=True)
```

# Primeira forma 
#### Histogramas

```python
for column in dataframe.columns:
    sns.histplot(dataframe[column], bins=20, kde=False, color='skyblue')
    
    plt.xlabel(column)
    plt.ylabel('Frec')
    plt.title(f'Histplot {column}')
    
    plt.show()
```

## QQplot

```python
for column in dataframe.columns:
    quantis_teoricos = stats.probplot(dataframe[column], dist="norm", fit=False)

    # Criar o Q-Q plot usando Seaborn e Matplotlib
    sns.set(style="whitegrid")
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=quantis_teoricos[0], y=quantis_teoricos[1])
    
    # Adicionar linha de referncia
    plt.plot([quantis_teoricos[0].min(), quantis_teoricos[0].max()], [quantis_teoricos[0].min(), quantis_teoricos[0].max()], color='red', linestyle='--')
    
    # Adicionar rtulos e ttulo
    plt.xlabel('Quantis Tericos')
    plt.ylabel('Quantis Observados')
    plt.title(f'Q-Q Plot - {column}')
```

### Teste de shapiro

```python
for column in dataframe.columns:
    stat, p_valor = shapiro(dataframe[column])
    print(f'{column}')
    print('--------------------------------------------------------------------')
    print(f"Estatstica do teste: {stat}")
    print(f"Valor-p: {p_valor}")
    print('--------------------------------------------------------------------')
    if p_valor > 0.05:
        print("No rejeitamos a hiptese nula. Os dados podem seguir uma distribuio normal.")
    else:
        print("Rejeitamos a hiptese nula. Os dados no seguem uma distribuio normal.")
    print('--------------------------------------------------------------------')
```

```python
from scipy.stats import kstest

for column in dataframe.columns:
    print(column)
    stat, p_valor = kstest(dataframe[column], 'norm')
    print('--------------------------------------------------------------------')
    print(f"Estatstica do teste: {stat}")
    print(f"Valor-p: {p_valor}")
    print('--------------------------------------------------------------------')
    # Interpretar o resultado
    if p_valor > 0.05:
        print("No rejeitamos a hiptese nula. Os dados podem seguir uma distribuio normal.")
    else:
        print("Rejeitamos a hiptese nula. Os dados no seguem uma distribuio normal.")
    print('--------------------------------------------------------------------')
```
