```python
import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
import seaborn as sns
```

```python
dataframe = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/iris.data')
```

```python
X = dataframe.drop(columns='Name')
y = dataframe['Name']
```

```python
tsne = TSNE(n_components=2)
```

```python
df_transform = tsne.fit_transform(X)
```

```python
coluns_name  = {0:'t_sne_1',1:'t_sne_2'}
df_transform = pd.DataFrame(df_transform)
df_transform.rename(columns=coluns_name,inplace=True)
```

```python
df_transform['target'] = y
```

```python
df_transform
```

```python
sns.scatterplot(x='t_sne_1', y='t_sne_2', hue='target', data=df_transform, palette='viridis', s=70)
```
