```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, power_transform

import plotly.express as px
from plotly.subplots import make_subplots
```

### Transformações matematicas 

1) Standard Scaling
2) Min Max Scaling
3) Logaritmicas
4) Box-Cox
5) Yeo-Johnson

```python
df = pd.read_csv('../../data/clean_dataset.csv')
df
```

```python
df_scaller = df[['debt','credit_score','income','years_employed']]
columns = df_scaller.columns.to_list()
df_scaller.head()
```

```python
scaler = StandardScaler()
df_scaller = scaler.fit_transform(df_scaller)
```

```python
df_scaller = pd.DataFrame(df_scaller,columns=columns)
df_scaller['approved'] = df['approved']
```

### Min Max Scaler

```python
df_min_max = df[['debt','credit_score','income','years_employed']]
columns = df_min_max.columns.to_list()
df_min_max.head()
```

```python
scaler = MinMaxScaler()
df_min_max = scaler.fit_transform(df_scaller)
```

```python
df_min_max = pd.DataFrame(df_min_max,columns=columns)
df_min_max['approved'] = df['approved']
df_min_max 
```

### Logaritmica

```python
df_log = df[['debt','credit_score','income','years_employed']]
columns = df_log.columns.to_list()
df_log.head()
```

```python
df_log = df_log+1
```

```python
df_log = df_log.apply(np.log10)
```

```python
df_log['approved'] = df['approved']
df_log 
```

### Potenciais Box-Cox 

```python
df_box_cox = df[['debt','credit_score','income','years_employed']]
df_box_cox = df_box_cox.loc[(df_box_cox >= 1).all(axis=1)]
columns = df_log.columns.to_list()
df_box_cox.reset_index(inplace=True,drop=True)
```

```python
columns.remove('approved')
```

```python
df_box_cox = power_transform(method='box-cox',X=df_box_cox)
```

```python
df_box_cox
```

```python
df_box_cox = pd.DataFrame(df_box_cox,columns=columns)
df_box_cox
```

### yeo_johnson

```python
df_yeo_johnson = df[['debt','credit_score','income','years_employed']]
columns = df_yeo_johnson.columns.to_list()
df_yeo_johnson.head()
```

```python
df_yeo_johnson = power_transform(method='yeo-johnson',X=df_box_cox)
```

```python
df_yeo_johnson = pd.DataFrame(df_yeo_johnson,columns=columns)
df_yeo_johnson['approved'] = df['approved']
df_yeo_johnson 
```
