```python
import pandas as pd
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL
```

```python
df = pd.read_csv('../../data/DailyDelhiClimateTrain.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
df
```

## Plot data 

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df.index, y=df.meantemp, mode='lines', name='Temperatures'))
fig.add_trace(go.Scatter(x=df.index, y=df.meanpressure, mode='lines', name='Presure'))
fig.add_trace(go.Scatter(x=df.index, y=df.humidity, mode='lines', name='Humidity'))
fig.add_trace(go.Scatter(x=df.index, y=df.wind_speed, mode='lines', name='Wind speed'))

fig.show()
```

```python
for column in df.columns:  
    df_colum = df[column]
    res = STL(df_colum).fit()
    res.plot()
    plt.show()
```
